This notebook explores the Netflix dataset from Kaggle. We load, clean, and prepare the data for future analysis.


```python
import pandas as pd

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Show first 5 rows
df.head()
```

```python
# Shape of dataset (rows, columns)
df.shape

# Overview of data
df.info()

# Summary statistics (for numeric fields)
df.describe(include='all')
```

```python
# Step 1: Rename column headers
df.columns = [col.lower().strip().replace(' ', '_') for col in df.columns]

# Step 2: Check for duplicate rows
print("Duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Step 3: Fill missing values
df['director'].fillna('Not Available', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Not Given', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
df['duration'].fillna('Not Available', inplace=True)

# Step 4: Convert 'date_added' to datetime (errors='coerce' will convert invalid to NaT)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Step 5: Strip whitespace and standardize case in text columns
df['type'] = df['type'].str.lower().str.strip()
df['country'] = df['country'].str.strip()

# Step 6: Save cleaned dataset
df.to_csv("netflix_cleaned.csv", index=False)

print("‚úÖ Data cleaned and saved successfully.")
```

## ‚úÖ Conclusion
The Netflix dataset has been successfully cleaned.  
- Handled missing values
- Checked and removed duplicates  
- Ready for further analysis or visualization.
